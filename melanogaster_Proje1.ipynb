```python
#kategorik değişkenlerimizi numerik olarak temsil edilen değişkenlere çeviriyoruz.

new_raw_data = pd.get_dummies(raw_data, columns = ['Season', 'Month'])
new_raw_data.head()
```

```python
#Feature scaling yapıyoruz. Böylece kodumuz daha hızlı ve optimize çalışacak.

scale_vars = ['meanTemp', 'maxTemp', 'minTemp', 'meanHumidity', 'maxHumidity', 'minH', 'meanPrecipitation', 'maxPrecipitation']
scaler = MinMaxScaler()
new_raw_data[scale_vars] = scaler.fit_transform(new_raw_data[scale_vars])
new_raw_data.head()
```

```python
#Hold-out yöntemi ile datamızı %90'ı train datası, %10'u test datası olacak biçimde ikiye bölüyoruz.
new_raw_data.drop('CS', inplace=True, axis=1)
X = new_raw_data.drop('CS1', axis=1).values # Girdiler
y = new_raw_data['CS1'].values # Predict edilmesi istenen vektör (yani CS değeri)

# X ve y olarak böldüğümüz data'nın boyutlarını elde ediyoruz.
print('X shape: {}'.format(np.shape(X)))
print('y shape: {}'.format(np.shape(y)))

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.9, test_size=0.1, random_state=0)
```

```python
# Decision tree'mizi böldüğümüz dataya fit ediyoruz.

dt = DecisionTreeClassifier(criterion='entropy', max_depth=2, random_state=1)
dt.fit(X_train, y_train)
```
